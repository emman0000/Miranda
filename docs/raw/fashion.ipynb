```python
styles = styles[
    styles["masterCategory"].isin([
        "Apparel",
        "Footwear"
    ])
]
```

```python
print(len(styles))
```

```python
styles = styles.dropna()
```

```python
styles = styles.drop_duplicates(subset="id")
```

```python
styles["articleType"].value_counts().head(25)
```

```python
#Curated List of categories
selected_articles = [
    # Tops
    "Tshirts",
    "Shirts",
    "Tops",
    "Kurtas",
    "Kurtis",
    "Tunics",
    "Sweatshirts",
    "Sweaters",
    "Jackets",

    # Bottoms
    "Jeans",
    "Trousers",
    "Shorts",
    "Track Pants",

    # Dresses
    "Dresses",
    "Sarees",

    # Footwear
    "Casual Shoes",
    "Sports Shoes",
    "Formal Shoes",
    "Heels",
    "Flats",
    "Sandals",
    "Flip Flops"
]

styles = styles[styles["articleType"].isin(selected_articles)]

print("Remaining products:", len(styles))
print()
print(styles["articleType"].value_counts())
```

Introducing Capping on items

```python
caps = {
    "Tshirts": 500,
    "Shirts": 500,
    "Casual Shoes": 500,
    "Sports Shoes": 500,
    "Kurtas": 500,
    "Tops": 500,
    "Heels": 400,
    "Flip Flops": 300,
    "Sandals": 300,
    "Formal Shoes": 300,
    "Jeans": 500,
    "Shorts": 400,
    "Trousers": 400,
    "Flats": 300
}

balanced = []

for article, group in styles.groupby("articleType"):
    if article in caps:
        balanced.append(
            group.sample(
                n=min(len(group), caps[article]),
                random_state=42
            )
        )
    else:
        balanced.append(group)

styles_balanced = pd.concat(balanced)

print(len(styles_balanced))
```

```python
styles_balanced["articleType"].value_counts()
```

```python
import os

image_folder = "fashion-dataset/images"

styles_balanced["image_exists"] = styles_balanced["id"].apply(
    lambda x: os.path.exists(f"{image_folder}/{x}.jpg")
)

print(styles_balanced["image_exists"].value_counts())
```

```python
styles_balanced = styles_balanced[
    styles_balanced["image_exists"]
].drop(columns="image_exists")
```

```python
import os

os.makedirs("miranda_dataset", exist_ok=True)

styles_balanced.to_csv(
    "miranda_dataset/styles.csv",
    index=False
)
```

```python
import shutil
from tqdm import tqdm

source = "fashion-dataset/images"
destination = "miranda_dataset/images"

os.makedirs(destination, exist_ok=True)

for image_id in tqdm(styles_balanced["id"]):
    shutil.copy(
        f"{source}/{image_id}.jpg",
        f"{destination}/{image_id}.jpg"
    )
```
